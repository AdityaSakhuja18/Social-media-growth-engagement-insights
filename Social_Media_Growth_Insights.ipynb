```python
import pandas as pd
```

```python
df = pd.read_csv("Social_Media_Analysis.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["total_engagement"] = df["likes"] + df["comments"] + df["shares"]
df.head()
```

```python
df.groupby("platform")["total_engagement"].sum()
```

```python
platform_rank = df.groupby("platform")["total_engagement"].sum().sort_values(ascending=False)
platform_rank
```

```python
df.groupby("platform")["total_engagement"].mean().sort_values(ascending=False)
```

```python
top_posts = df.sort_values("total_engagement", ascending=False).head(5)
top_posts
```
